# Q.1 Import the libraries




```python
import numpy as np
import pandas as nd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
%matplotlib inline
```

# Q.2 Get the Dataset

```python
dt=nd.read_csv('tips.csv')
dt
```

# Q3. Plot the distribution of total_bill in the given dataset.

```python


sns.distplot(dt['total_bill'],kde=False,rug=True);
```

# Q.4 - Plot the destribution plot for tips.

```python
sns.distplot(dt['tip'],kde=False);
```

# Q5. Make a joint plot between tip and total bill.

```python
sns.jointplot(dt['total_bill'],dt['tip'])
```

# Q6. Give additional attribute kind = "hex" in the above plot.

```python
sns.jointplot(dt['total_bill'],dt['tip'],kind="hex");
```

# Q7. Make pairplot between total bill and tip.

```python

sns.pairplot(dt[['total_bill','tip']]);
```

# Q.8 -  Make striplot between sex and tip.

```python
sns.stripplot(dt['sex'],dt['tip'])
```

# Q.10 Make bar plot with day and total bill.

```python
sns.barplot(dt['day'],dt['total_bill'])
```

# Q.11 - Make a plot to count total males and females for each day.

```python
sns.countplot(dt['day'],hue=dt['sex'])
```

# Q.12 -  Use factorplot to plot multiple categorical variables like day, smoker, peoplecount.

```python
sns.catplot(x="day",y='size',hue='sex',col='smoker',data=dt,kind="bar")
```
